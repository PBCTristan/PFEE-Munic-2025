```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from fourier import remove_noise_fft
```

```python
signal_length = 1
t = np.arange(0, signal_length, 0.001)
f = np.sin(2 * np.pi * t * 42) + np.cos(2 * np.pi * t * 69)
noisy_f = f + 2.5 * np.random.randn(len(t))

# Printing the funcs

fig, axis = plt.subplots(2, 1)

plt.sca(axis[0])
plt.plot(t, f)
plt.xlim(t[0], t[-1])

plt.sca(axis[1])
plt.plot(t, noisy_f)
plt.xlim(t[0], t[-1])
```

```python
deaf_f = remove_noise_fft(noisy_f, signal_length)

###

plt.clf()
plt.plot(t, deaf_f)
plt.xlim(t[0], t[-1])
```
